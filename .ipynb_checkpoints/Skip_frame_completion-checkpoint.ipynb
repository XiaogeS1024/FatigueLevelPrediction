```python
import pandas as pd
import numpy as np
import math

from warnings import simplefilter

    
simplefilter(action='ignore') #忽略警告

def insert(df, i, df_add):
    # 指定第i行插入一行数据
    df1 = df.iloc[:i, :]
    df2 = df.iloc[i:, :]
    df_new = pd.concat([df1, df_add, df2], ignore_index=True)
    return df_new

level_now=0
index_out=1

for level in range(0,3):
    if level == 0 :
        index_max = 215
    if level == 1 :
        index_max = 100
    if level == 2 :
        index_max = 144
    for index in range(1,index_max+1):
        file_name="data/level_"+str(level)+"/level_"+str(level)+" ("+str(index)+").csv";
        print(file_name)
        data = pd.read_csv(file_name,encoding = 'gb2312',names=['id', 'frame', 'x', 'y'])  # 读入csv文件中的特征
        #如果文件不空
        if data.shape[0] != 1:
            #判断缺失的帧数
            frame_first=0;
            frame_count=0;
            frame_index=-1;
            length=data.shape[0];
            frame_first=int(data.iloc[1]["frame"]);
            for i in range(1,length-1):
                #帧数一般按最后一帧编号算
                frame=int(data.iloc[i]["frame"]);
                if frame!=frame_index:
                    frame_index=frame;
                    frame_count=frame_count+1;
            lost_frame=frame-frame_count;
    #         print(lost_frame)
    #         不废弃的情况
            if lost_frame < 10:
                ##当第一帧不存在时，frame值需要集体减frame_first再加1              
                if frame_first != 1:                    
                    df_add=data[0:1]
                    data.drop([0],inplace = True)
                    data['frame']=data['frame'].astype(int)
                    data['frame']=data['frame'].map(lambda x : x-frame_first+1)        
                    data = insert(data, 0, df_add)
                    #帧数也要重新算
                    frame=int(data.iloc[length-10]["frame"]); 

                
                #没有丢帧就直接输出        
                if lost_frame == 0:                    
                    data.drop([0],inplace = True)
                    data['frame']=data['frame'].astype(int)
                    data['id']=data['id'].astype(int)
                    data['x']=data['x'].astype(int)
                    data['y']=data['y'].astype(int)
                    df_end = data
                #丢帧的话处置
                else:
                    length=data.shape[0];
                    df=[]
                    frame=0
                    df_end=pd.DataFrame(columns = ["id", "frame", "x", "y"]) 
                    k=0

                    # 创建68个dataframe
                    for i in range(0,68):
                        df_a_row= pd.DataFrame(columns = ["id", "frame", "x", "y"]) 
                        df.append(df_a_row)
                    # print(df[0])  #是个空dataframe

                            #插值补帧

                        #划分为68个dataframe
                    for i in range(0,length-1):
                        i=i+1;
                        df[int(data.iloc[i]['id'])].loc[int(data.iloc[i]['frame'])]=[data.iloc[i]['id'], data.iloc[i]['frame'],data.iloc[i]['x'],data.iloc[i]['y']];
                        frame=int(data.iloc[i]["frame"]);
                        #frame即为帧数

                        #df_full是满帧的帧那一列
                    temp_array = np.arange(1, frame+1).reshape(frame,1)
                    df_full = pd.DataFrame(temp_array,columns=list('f'))
                    df_full.rename(columns={'f':'frame'},inplace=True)
                    df_full['frame']=df_full['frame'].astype(int)

    #                 print(df[2]['x'])        
                    for i in range(0,68):  
                        #先做类型转换
                        df[i]['frame']=df[i]['frame'].astype(int)
                        df[i]['id']=df[i]['id'].astype(int)
                        df[i]['x']=df[i]['x'].astype(int)
                        df[i]['y']=df[i]['y'].astype(int)
                        #合并dataframe
                        df[i] = pd.merge(df[i], df_full, on='frame', how='outer').sort_values('frame')
                        #做插值
                        df[i]['id'] = i
                        df[i]['x'] = df[i]['x'].interpolate(method='linear')
                        df[i]['y'] = df[i]['y'].interpolate(method='linear')
                        #再转一遍int，目前是小数状态
                        df[i]['frame']=df[i]['frame'].astype(int)
                        df[i]['id']=df[i]['id'].astype(int)
                        df[i]['x']=df[i]['x'].astype(int)
                        df[i]['y']=df[i]['y'].astype(int)
                    for i in range(0,frame):
                        for j in range(0,68):
                            df_end.loc[k] = [df[j].iloc[i]['id'],df[j].iloc[i]['frame'],df[j].iloc[i]['x'],df[j].iloc[i]['y']]
                            k=k+1

                    #df_end即为所求文件  

                #输出到文件

                if level != level_now:
                    level_now=level;
                    index_out=1;

                file_name_new="data_Skip_frame_completion/level_"+str(level)+"/"+str(level)+"_"+str(index_out)+".csv";
                df_end.to_csv(file_name_new,index=False,sep=',')
                print(file_name_new)
                index_out=index_out+1;
```
