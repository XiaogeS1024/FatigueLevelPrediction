```python
import pandas as pd
import numpy as np
import math
from warnings import simplefilter

simplefilter(action='ignore') #忽略警告


for level in range(0,3):
    if level == 0 :
        index_max = 199
    if level == 1 :
        index_max = 90
    if level == 2 :
        index_max = 132
    for index in range(1,index_max+1):
        file_name="data_Skip_frame_completion/level_"+str(level)+"/"+str(level)+"_"+str(index)+".csv";
        print(file_name)
        data = pd.read_csv(file_name,encoding = 'gb2312',names=['id', 'frame', 'x', 'y'])  # 读入csv文件中的特征
                 
        length=data.shape[0]
        center={}
        frame_col={}
        for i in range(0,length-1):
            i=i+1;
            id=data.iloc[i]["id"];
            frame=int(data.iloc[i]["frame"]);
            if id == "30" :
                center[frame]=float(data.iloc[i-2]["y"])
#                 print(center[frame])
             
        for k in center:
#             print(k)
            frame_col[k]=k;

        df = pd.DataFrame({'frame':frame_col,'center':center})
 
        file_name_new="center_point/level_"+str(level)+"/"+str(level)+"_"+str(index)+".csv";
        df.to_csv(file_name_new,index=False,sep=',')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
%matplotlib inline

fig= plt.figure()
ax = fig.add_axes([0,0,1,1]) 
ax.set_ylim(50, 600)
ax.set_xlim(0, 90)
ax.set_xticks([0,20,40,60,80])

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# #matplotlib画图中中文显示会有问题，需要这两行设置默认字体

plt.title('level_0')
ax.set_xlabel("帧号")
ax.set_ylabel("中心点坐标")

for index in range(1,50):
        file_name="center/level_0/0_"+str(index)+".csv";
#         print(file_name)
        data = pd.read_csv(file_name,encoding = 'gb2312',names=['frame', 'center'])  # 读入csv文件中的特征
        length=data.shape[0];
        d=[]
        x=[]
        for i in range(1,length):
            x.append(i)
            d.append(float(data.iloc[i]["center"]))
#         print(d)
        ax.plot(x,d)   
```
