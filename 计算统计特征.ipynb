```python
import pandas as pd
import numpy as np
import math
from warnings import simplefilter

simplefilter(action='ignore') #忽略警告


k=0 # 标记记录行数
file_index={};


eye_count_radio={};#闭眼帧的比例
MOA_count_radio={};#打哈欠帧的比例
center_std={};
level_label={};
        
        
for level in range(0,3):
    if level == 0 :
        index_max = 199
    if level == 1 :
        index_max = 90
    if level == 2 :
        index_max = 132
     
 
    for index in range(1,index_max+1):
        head=[];
        file_name="data_final_2/level_"+str(level)+"/"+str(level)+"_"+str(index)+".csv";
#         print(file_name)
        data = pd.read_csv(file_name,encoding = 'gb2312',names=['frame','center','eyes_ratio','MOA'])  # 读入csv文件中的特征

        length=data.shape[0]
        
        eye_count=0;#闭眼帧数
        MOA_count=0;#打哈欠帧数
        
        for i in range(1,length):
            
            head.append(float(data.iloc[i]["center"]))
            
            #眼睛睁开比例小于20%算闭眼
            if float(data.iloc[i]["eyes_ratio"]) < 0.2:
                eye_count+=1
            
            #嘴巴张合度大于60°算打哈欠，即MOA小于0.5
            if float(data.iloc[i]["MOA"]) < 0.5:
                MOA_count+=1
        
        #计算中心点的位置的标准差
        center_std[k] = np.std(head)
        print(center_std[k])
    
        file_index[k]=k
        eye_count_radio[k]=eye_count/70.0;#闭眼帧的比例
        MOA_count_radio[k]=MOA_count/70.0;#打哈欠帧的比例
        level_label[k]=level;
        k+=1
    
df = pd.DataFrame({'file_index':file_index,
                    'center_std':center_std,
                    'eye_count_radio':eye_count_radio,
                    'MOA_count_radio':MOA_count_radio,
                    'level_label':level_label})
 
file_name_new="data_2.csv";
df.to_csv(file_name_new,index=False,sep=',')
```
