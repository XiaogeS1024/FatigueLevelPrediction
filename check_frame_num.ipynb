```python
import pandas as pd
import numpy as np
import math
from warnings import simplefilter


simplefilter(action='ignore') #忽略警告

for level in range(0,3):
    if level == 0 :
        index_max = 215
    if level == 1 :
        index_max = 100
    if level == 2 :
        index_max = 144  
    file={};
    frame_count_arr={};
    frame_last_arr={};
    lost_frame_count_arr={};
    for index in range(1,index_max+1):
        file_name="data/level_"+str(level)+"/level_"+str(level)+" ("+str(index)+").csv";
        
        data = pd.read_csv(file_name,encoding = 'gb2312',names=['id', 'frame', 'x', 'y'])  # 读入csv文件中的特征
        print(file_name)
        frame_count=0;
        frame_index=-1;
        length=data.shape[0];
        for i in range(1,length):
            frame=int(data.iloc[i]["frame"]);
            if frame!=frame_index:
                frame_index=frame;
                frame_count=frame_count+1;
        
        file[index]=index;
        frame_count_arr[index]=frame_count;
        frame_last_arr[index]=frame;
        lost_frame_count_arr[index]=frame-frame_count;
        
                
    df = pd.DataFrame({'file':file,'frame_count':frame_count_arr,
                          'frame_last':frame_last_arr,"lost_frame_count":lost_frame_count_arr})
 
    file_name_new="frame_count/level_"+str(level)+"/frame_count.csv";
    df.to_csv(file_name_new,index=False,sep=',')
#         print("该数据集的总帧数为：",frame_count)
#         print("最后一帧的帧号为：",frame)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


frame_count=[];
file=0;

for level in range(0,3):
    file_name="frame_count/level_"+str(level)+"/frame_count.csv";
    data = pd.read_csv(file_name,encoding = 'gb2312',names=['file', 'frame_count','frame_last','lost_frame_count'])  # 读入csv文件中的特征
    length=data.shape[0];
    for i in range(1,length):
        file=file+1;
        frame_count.append(int(data.iloc[i]["lost_frame_count"]));

        
# plt.rcParams['font.sans-serif']=['SimHei']
# plt.rcParams['axes.unicode_minus'] = False
# #matplotlib画图中中文显示会有问题，需要这两行设置默认字体

plt.figure(figsize=(10, 10))
plt.xlabel('文件编号')
plt.ylabel('缺帧数')
plt.xlim(xmax=500,xmin=0)
plt.ylim(ymax=90,ymin=0)

area = np.pi * 4**2  # 点面积 
plt.scatter(range(459),frame_count)

plt.legend()
plt.title('缺帧数一览图')
plt.yticks(np.arange(0, 90, step=2))
plt.show()
     
        
```
