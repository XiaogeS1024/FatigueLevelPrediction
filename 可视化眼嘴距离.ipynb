```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
%matplotlib inline

fig= plt.figure()
ax = fig.add_axes([0,0,1,1]) 
ax.set_ylim(20, 120)
ax.set_yticks([20,40,60,80,100,120])
ax.set_xlim(0, 90)
ax.set_xticks([0,20,40,60,80])

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# #matplotlib画图中中文显示会有问题，需要这两行设置默认字体

plt.title('level_0')
ax.set_xlabel("帧号")
ax.set_ylabel("眼-嘴距离")

for index in range(1,50):
        file_name="eye_mouth_ditance/level_0/0_"+str(index)+".csv";
#         print(file_name)
        data = pd.read_csv(file_name,encoding = 'gb2312',names=['frame', 'eye_mouth_distance'])  # 读入csv文件中的特征
        length=data.shape[0];
        d=[]
        x=[]
        for i in range(1,length):
            x.append(i)
            d.append(float(data.iloc[i]["eye_mouth_distance"]))
#         print(d)
        ax.plot(x,d)   
```
